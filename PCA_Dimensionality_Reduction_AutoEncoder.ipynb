```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(100,3,2,random_state=101)
```

```python
print(data)
```

```python
%matplotlib inline
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaledData = scaler.fit_transform(data[0])
```

```python
datax = scaledData[:,0]
datay = scaledData[:,1]
dataz = scaledData[:,2]
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(datax,datay,dataz,c=data[1])
```

```python
from tensorflow import keras as k
```

```python
model = k.models.Sequential([k.layers.Dense(3),k.layers.Dense(2,name="hid"),k.layers.Dense(3)])
```

```python
model.compile(optimizer='rmsprop',metrics=['acc'],loss='mean_squared_error')
```

```python
his = model.fit(scaledData,scaledData,epochs=1000)
```

```python
print(his)
```

```python
l = model.get_layer(name="hid")
```

```python
we = l.get_weights()[0]
l.get_config()
```

```python
x = scaledData@we + l.get_weights()[1]
```

```python
x
```

```python
plt.scatter(x[:,0],x[:,1],c=data[1])
```
